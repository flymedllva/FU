4.2 Посчитайте средний рейтинг к каждому из рецептов. Для скольких рецептов отсутствуют отзывы?

```python
print(f'Количество рецептов без отзывов: {reviews.review.isna().sum()}')
```

```python
recipec_rating_mean = dict()
recipes_id = reviews.recipe_id.unique()
for i in range(len(recipes_id)):
    recipec_rating_mean[recipes_id[i]] = round(reviews[reviews.recipe_id == recipes_id[i]].rating.mean(),2)
    print(f'Средний рейтинг для рецепта {recipes_id[i]}: {round(reviews[reviews.recipe_id == recipes_id[i]].rating.mean(),2)}')
```

### Объединение таблиц `pd.DataFrame`

5.1 При помощи объединения таблиц, создайте `DataFrame`, состоящий из четырех столбцов: `id`, `name`, `user_id`, `rating`. Рецепты без отзывов должны отсутствовать в данной таблице. Подтвердите правильность работы вашего кода, выбрав рецепт, не имеющий отзывов, и выведя на экран строку из полученного `DataFrame`, содержащую информацию об этом отзыве.

```python
not_null = reviews[reviews.review.notnull()][['user_id','recipe_id','rating']]
not_null[not_null.recipe_id == 286]

recipes_id = np.array(reviews.recipe_id.value_counts())
recipes_id_notnull = np.array(not_null.recipe_id.value_counts())

#смотрим, какие рецепты без отзывов
set(recipes_id) - set(recipes_id_notnull)
```

```python
# pd.merge(product,customer,left_on='Product_name',right_on='Purchased_Product')
table1 = pd.merge(recipes[['id','name']],reviews[reviews.review.notnull()][['user_id','recipe_id','rating']],left_on='id',right_on='recipe_id').drop(['recipe_id'],axis=1)
table1.head()
```

5.2 При помощи объединения таблиц и группировок, создайте `DataFrame`, состоящий из трех столбцов: `recipe_id`, `name`, `review_count`. У рецептов, для которых отсутствуют отзывы, в соответствущем столбце должен быть указан 0. Подтвердите правильность работы вашего кода, выбрав рецепт, не имеющий отзывов, и выведя на экран строку из полученного `DataFrame`, содержащую информацию об этом отзыве.


```python
reviews.groupby('recipe_id').count()[['review']].reset_index(level=0)
```

```python
reviews.head()
```

```python
table2 = recipes[['id','name']].merge(reviews.groupby('recipe_id').count()[['review']].reset_index(level=0), left_on='id',right_on='recipe_id', how='left').fillna(0, downcast='infer').drop(['recipe_id'],axis=1)
```

### Сохранение таблиц `pd.DataFrame`

6.1 Отсортируйте таблицу в порядке убывания величины столбца `name_word_count` и сохраните результаты выполнения заданий 3.1-3.3 в csv файл. 

```python
recipes['name_word_count'] = recipes['name'].str.split().str.len()
recipes = recipes.sort_values(by=['name_word_count'],ascending=False)
recipes.head()
```
